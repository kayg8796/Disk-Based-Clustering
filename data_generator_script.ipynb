```python
import pandas as pd
#df_tgenome = pd.read_csv('genome-tags.csv')
```

```python
df_genre = pd.read_csv('movies.csv')#for genre as it is
df_ratings = pd.read_csv('ratings.csv')#for ratings with mordification
df_tags  = pd.read_csv('tags.csv') #for tags with mordification
df_relevance = pd.read_csv('genome-scores.csv')
dft_genome =  pd.read_csv('genome-tags.csv')
```

```python
#modelling dataset for genre
```

```python
df_movies = df_genre.set_index('movieId')
df_movies.genres=df_movies.genres.apply(lambda x: x.replace('|',' '))
df_movies.title=df_movies.title.apply(lambda x: x.replace(',',' '))    #removing commas since they disturb importation
df_movies.head(20)
```

```python
genre_csv=df_movies.to_csv('movie_genre.csv',header=False,index=False)
```

```python
#modelling dataset for ratings
```

```python
df_ratings.head()
```

```python
dfr_new=df_ratings.groupby(['movieId','userId']).rating.mean().unstack().fillna(0)
ratings = pd.merge(df_movies,dfr_new,left_index=True,right_index=True,how='inner').drop(columns='genres')
ratings.head()
```

```python
ratings_csv=ratings.to_csv('movie_ratings.csv',header=False,index=False)
```

```python
#modelling dataset for tagging
```

```python
df_rel=df_relevance[df_relevance['relevance']> 0.5]
df_rel = pd.merge(df_rel,dft_genome,on='tagId',how = 'inner')
dftag=df_rel.groupby('movieId').tag.agg(lambda x : ' '.join([str(k) for k in x]))
dftn = pd.merge(df_movies, dftag, left_index=True, right_index=True, how='inner').drop(columns='genres')
dftn.title=dftn.title.apply(lambda x: x.replace(',',' ')) 
dftn.tag=dftn.tag.apply(lambda x: x.replace(',',' '))
dftn.head(20)
```

```python
tags_csv=dftn.to_csv('movie_tags.csv',header=False,index=False)
```

```python
#modelling dataset for combined distance
#dftn
```

```python
dfc = pd.merge(df_movies, dftag, left_index=True, right_index=True, how='inner')
df_combine = pd.merge(dfc,dfr_new,left_index=True,right_index=True, how='inner')
#df_combine.iloc[:,3:] = df_combine.fillna(0)
#df_combine.iloc[:,:3] = df_combine.iloc[:,:3].fillna('0')
df_combine.tag=df_combine.tag.apply(lambda x: x.replace('[^\w\s]','')) #removing punctuations
df_combine.head(3)
```

```python
combine_csv=df_combine.to_csv('combine.csv',header=False,index=False)
```
